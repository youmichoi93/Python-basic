```python
# [a-zA-Z] : 알파벳 모두
# [0-9 ] : 숫자

import re
alp = re.compile('[a-zA-z]')
num = re.compile('[0-9]')

a = input()

if alp.match(a):
  print("문자")
elif num.match(a):
  print("숫자")
else:
  print("기호")
```

```python
import re
alp = re.compile('[a-zA-z]')
num = re.compile('[0-9]')

a = input()
b = "" #문자들
c = "" #숫자들
d = "" #기호들

for i in range(len(a)):    # a = abcde,  5  -> 0,1,2,3,4
  if alp.match(a[i]):
    b = b+a[i]  #문자들
  elif num.match(a[i]):
    c = c+a[i]  #숫자들
  else:
    d = d+a[i]  #기호들

print("%s\n%s\n%s" % (b,c,d))

```

```python
import re
word = input()

st=re.compile('[a-zA-Z]')  #문자 찾음
n=re.compile('[0-9]')      #숫자 찾음
sp=re.compile('\W')        #둘다 아닌 것 찾음

print(''.join(st.findall(word)))  #문자출력
print(''.join(n.findall(word)))   #숫자출력
print(''.join(sp.findall(word)))  #나머지 출력
```
