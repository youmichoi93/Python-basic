```python
#게시판 페이징하기


def getTotalPage(m,n):
    if m % n ==0:
        return m //n
    else:
        return m // n + 1

print(getTotalPage(5,10))  # 5가 게시물의 갯수 10 한 페이지당 게시물 개수
print(getTotalPage(25,10))  # 10,10,5
print(getTotalPage(30,10)) #4 가 출력(딱 떨어짐), 따라서 0일때 따로 처리, if로 디버깅
```

```python
## 정규표현식 #########  파이썬 아니고 또 다른 언어임. 고급이라 그냥 개념만 이해.
```

```python
#파이썬으로 할때, 복잡함


data = '''
park 800905-1049118
kim 700905-1059119
'''


result = []
for line in data.split("\n"):
    word_result = []
    for word in line.split(' '):
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word = word[:6]+'-'+"*******"
        word_result.append(word)
    result.append(' '.join(word_result))
print('\n'.join(result))
```

```python
#정규식으로 할때


import re

data = '''
park 800905-1049118
kim  700905-1059119
'''

pat = re.compile("(\d{6})[-]\d{7}")
print(pat.sub("\g<1>-*******",data))
```
